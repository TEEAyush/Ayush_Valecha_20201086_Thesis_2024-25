This script will extract the names of the pollen allergen genes from the supplementary material tables from Castric et al. 2024. The names will then be stored in a csv file.

```python
import pandas as pd
```

```python
diff_gene = pd.read_excel(r"C:\Sylvain_Billiard\Masters_Thesis_Work\Preliminary Readings\Castric_et_al_2024\Castric et al 2024 suppl. tables.xlsx", sheet_name =["I","J"])
```

```python
diff_gene["I"].head()
```

```python
# drop the empty row

diff_gene["I"] = diff_gene["I"].drop([0])

# drop the empty column

diff_gene["I"] = diff_gene["I"].drop(columns = ["Unnamed: 0"])

diff_gene["I"].head()
```

```python
# make row 1 the new header
new_header = diff_gene["I"].iloc[0]
diff_gene["I"] = diff_gene["I"][1:]
diff_gene["I"].columns = new_header
```

```python
diff_gene["I"].head()
```

```python
diff_gene["J"].head()
```

```python
# repeat the same steps for sheet J

diff_gene["J"] = diff_gene["J"].drop([0]) # drop the empty row
diff_gene["J"] = diff_gene["J"].drop(columns = ["Unnamed: 0"]) # drop the empty column
new_header = diff_gene["J"].iloc[0] # make row 1 the new header
diff_gene["J"] = diff_gene["J"][1:]
diff_gene["J"].columns = new_header

diff_gene["J"].head()
```

```python
# slect the columns where the predicted function is " Pollen Ole e 1 allergen and extensin family protein"

pag_I = diff_gene["I"][diff_gene["I"]["Predicted function"] == ' Pollen Ole e 1 allergen and extensin family protein ']
pag_J = diff_gene["J"][diff_gene["J"]["Predicted function"] == ' Pollen Ole e 1 allergen and extensin family protein ']
```

```python
# store pag_I and pag_J in a new excel file

with pd.ExcelWriter(r"C:\Sylvain_Billiard\Masters_Thesis_Work\Preliminary Readings\Castric_et_al_2024\Pollen Ole e 1 allergen and extensin family protein.xlsx") as writer:
    pag_I.to_excel(writer, sheet_name = "I", index = False)
    pag_J.to_excel(writer, sheet_name = "J", index = False)

```
